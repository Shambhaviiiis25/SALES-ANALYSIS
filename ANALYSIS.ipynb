**REMOVE DUPLICATES IF ANY**

```python
df=df.drop_duplicates()
df
```

# ANALYSIS

**Question 1: What is the best month for sales in 2022? How much was earned that month?**

**Convert the 'Date' column to datetime format and create a new column Month.**

```python
#FORMAT CHANGE for dataframe df
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')
#CREATING A NEW COLUMN in datframe df
df['Month'] = df['Date'].dt.month_name()
#FORMAT CHANGE for dataframe df1
df1['Date'] = pd.to_datetime(df1['Date'], format='%d-%b-%y')
#CREATING A NEW COLUMN in datframe df1
df1['Month'] = df1['Date'].dt.month_name()
```

```python
df1.groupby("Month")["Sales"].sum() 
```

**Question 2: Which city had the highest number of sales?**

```python
df1.groupby("ship-state")["Sales"].sum().sort_values(ascending=False).head(10)
```

**QUESTION 3: What are the TOP performing Product categories?**

```python
df1.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

**QUESTION 4: What is the Average order Value for each month?**

```python
df1.groupby("Month").apply(lambda x: x["Sales"].sum() / x["Qty"].sum()).sort_values(ascending=False)
```

**QUESTION 5: What is the delivery status?**

```python
df1.groupby("Status")["Sales"].sum().sort_values(ascending=False)
```

**QUESTION 6: To find out if there is any correlation between Shipping speed and sales revenue?**

```python
df1['Expedited'] = (df1['ship-service-level'] == 'Expedited').astype(int)
df1['Standard'] = (df1['ship-service-level'] == 'Standard').astype(int)

# Calculating the correlation between Sales and the Expedited service level
correlation_expedited = df1['Sales'].corr(df1['Expedited'])

# Calculating the correlation between Sales and the Standard service level
correlation_standard = df1['Sales'].corr(df1['Standard'])

# Creating a DataFrame to hold the correlation values
correlation_df = pd.DataFrame({'Service Level': ['Expedited', 'Standard'],'Correlation with Sales': [correlation_expedited, correlation_standard]})

correlation_df
```

**QUESTION 7: To find out which product size gets highest number of rejections?**

```python
# Filtering the DataFrame for rows where 'Status' is 'Shipped - Rejected by Buyer'
rejected= df1[df1["Status"] == "Shipped - Rejected by Buyer"]
# Grouping by the 'Size' and counting the number of occurrences for each size
rejection_counts= rejected.groupby("Size").size().reset_index(name="Rejection Count")
rejection_counts
```

**QUESTION 8: To find out which product category gets highest number of Cancellations?**

```python
#Filtering
cancelled=df1[df1["Status"]=="Cancelled"]
#groupby()
cancellation_count=cancelled.groupby("Category").size().sort_values(ascending=False).reset_index(name="Cancelled Orders")
cancellation_count
```
